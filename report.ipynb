```python
def create_day_route_df(obj, day_num, vehicle):
    '''Вывод датасета по маршруту одного броневика.
    Входные параметры:
    obj: класс отработавшего оптимизатора
    day_num: номер дня от 1 до 91
    vehicle: броневик, значения из списка [0,1,2,3,4] '''
    
    y=[x+1 for x in obj.route_vehicle[day_num][vehicle]]
    dist_with_service = [int(obj.distance_matrix[y[i]][y[i+1]]) for i in range(len(y)-1)]
    dist= list(map(lambda x: max(0, x-10), dist_with_service))
    
    day_route_df = pd.DataFrame(index=obj.balance.iloc[obj.route_vehicle[day_num][vehicle][1:]].index,
                            columns=['vehicle_num','tid','start_time', 'end_time', 'route_time_to_tid',
                                     'route_time_to_tid_with_service'])
    day_route_df['vehicle_num'] = vehicle
    day_route_df['tid'] = day_route_df.index
    start_day = pd.to_datetime(obj.balance.iloc[:,day_num].name)
    day_route_df.loc[:,'route_time_to_tid'] = dist
    day_route_df.loc[:,'route_time_to_tid_with_service'] = dist_with_service
    day_route_df['route_time_to_tid_with_service_cumsum'] = day_route_df.route_time_to_tid_with_service.cumsum()
    day_route_df.loc[:,'start_time'] = start_day + day_route_df.route_time_to_tid_with_service_cumsum.apply(lambda x: 
                                            timedelta(hours=8, minutes=x))
    day_route_df['end_time'] = day_route_df['start_time'] + timedelta(minutes=obj.service_time)
    return day_route_df
```

```python
create_day_route_df(cl, 90, 0)
# расстояния в минутах округлены для удобства в бОльшую сторону (12.1 мин => 13 мин) 
```

__Условия расчета остатка по ТЗ (как считали мы вначале):__

    2.  Принимаем, что остаток в терминале увеличивается на сумму оборота в терминале разово. Т.е. утром известны остатки в терминалах и они не меняются за время работы бригад броневиков. Остатки на утро = остатки утра вчера  + обороты вчера.
Например, для терминала 406136 на утро 02.09 остаток = 250000, если 01.09 терминал не обслужили и = 90000, если терминал 01.09 обслужили.


__Условия расчета остатка по Отчету (пересчитаем сейчас):__

    Лист "остатки на конец дня" показывает сумму остатков в устройствах в разрезе дат на конец дня. Т.е. в случае, если устройство было инкассировано, в ячейке точно должен быть 0

```python
cor_balance=pd.DataFrame(data=0, index=cl.incomes.index, columns = cl.incomes.columns[1:])
cor_balance.insert(loc=0, column='остаток на 31.08.2022 (входящий)',
                            value=cl.incomes['остаток на 31.08.2022 (входящий)'])
```

```python
cash_rate_per_day_fact_df_cor = cor_balance.iloc[:,1:]
```

```python
%%time
d=1
while d < 92:
    no_visits = cl.day_df_dict[d].cost_per_service_fact.apply(lambda x: False if x < 0 else True)
    cor_balance.iloc[:,d][no_visits] = cor_balance.loc[no_visits].iloc[:,d-1] + cl.incomes.loc[no_visits].iloc[:,d]
    cash_rate_per_day_fact_df_cor.iloc[:,d-1][no_visits] = - cor_balance.loc[no_visits].iloc[:,d]*cl.p_per_day
    d+=1
    
```
